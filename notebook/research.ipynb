```python
import pandas as pd
import numpy as np
```

```python
# load books dataset

books = pd.read_csv("BX-Books.csv", sep=';', error_bad_lines=False, encoding='latin-1')
books

# image-url-S = small size image
# image-url-M = medium size image
# image-url-L = large size image
```

```python
books.iloc[200,-1] # Or books.iloc[200]["Image-URL-L"]
```

```python
books.shape
```

```python
books.columns
```

```python
books = books.drop(columns=["Image-URL-S", "Image-URL-M"])
books
```

```python
# rename columns name
books.rename(columns={
    "Book-Title":"title",
    "Book-Author":"author",
    "Year-Of-Publication":"year",
    "Publisher":"publisher",
    "Image-URL-L":"image_url"
}, inplace=True)
```

```python
books.sample(5)
```

```python
# load Users data set
users = pd.read_csv("BX-Users.csv", sep=';', error_bad_lines=False, encoding='latin-1')
users
```

```python
users.shape
```

```python
users.rename(columns={
    "User-ID":"user_id",
    "Location":"location",
    "Age":"age"
}, inplace=True)
```

```python
users.sample(5)
```
