```python
# load rating dataset

ratings = pd.read_csv("BX-Book-Ratings.csv", sep=';', encoding='latin-1', error_bad_lines=False)
ratings
```

```python
ratings.shape
```

```python
ratings.rename(columns={
    "User-ID":"user_id",
    "Book-Rating":"rating"
}, inplace=True)
```

```python
ratings.sample(5)
```

```python
books.shape, users.shape, ratings.shape
```

```python
# users who ranting more than 200 books 
x = ratings["user_id"].value_counts()>200
```

```python
x
```

```python
x[x].shape
```

```python
# user-id of that people
y = x[x].index
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
ratings
```

```python
ratings_with_books = ratings.merge(books, on='ISBN')
ratings_with_books
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
number_rating
```

```python
number_rating.rename(columns={'rating':'num_of_rating'}, inplace=True)
number_rating
```

```python
final_rating = ratings_with_books.merge(number_rating, on='title')
final_rating
```

```python
# Let's take those books which got at least 50 rating of user

final_rating = final_rating[final_rating['num_of_rating']>=50]
final_rating
```

```python
final_rating[['title', 'user_id']].value_counts()
```

```python
# drop duplicates of title with same user_id 
# means, same user_id gives more than one rating on same book
final_rating.drop_duplicates(['user_id', 'title'], inplace=True)
final_rating
```

```python
final_rating.isnull().sum()
```

```python
# let's create pivot table
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
book_pivot
```

```python
book_pivot.fillna(0.0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
type(book_sparse)
```

```python
# NearestNeighbors use here for clustering
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```
